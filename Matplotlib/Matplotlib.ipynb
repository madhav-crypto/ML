```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
heart = pd.read_csv('heart.csv')
heart.head()
```

```python
heart.plot.hist(subplots=True,figsize=(10,30));
```

```python
over50 = heart[heart['age']>50]
over50
```

```python
over50.plot(kind='scatter',
           x='age',
           y='chol',
            c='target',
           figsize=(10,6))
```

```python
fig,ax = plt.subplots(figsize=(10,6))

over50.plot(kind='scatter',
           x='age',
           y='chol',
           c='target',
           ax=ax);
```

```python
plt.style.available
```

```python
plt.style.use('seaborn')
```

```python
fig,ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x= over50['age'],
          y = over50['chol'],
          c = over50['target'])

ax.set(title='Heart disease chol by age',xlabel='age',ylabel='chol')
ax.axhline(over50['chol'].mean())
ax.legend(*scatter.legend_elements(),title='target');
```

```python
over50['chol'].mean()
```

```python
fig, (ax0,ax1) = plt.subplots(nrows=1,ncols=2,
                             figsize=(15,5))

scatter0 = ax0.scatter(x=over50['age'],
            y=over50['chol'],
           c = over50['target'],
                      cmap='winter')

scatter1= ax1.scatter(x=over50['age'],
           y=over50['thalach'],
           c=over50['target'],
                     cmap='winter')

ax0.set(title='age and cholestrol',xlabel='age',ylabel='cholestrol')
ax1.set(title='age and highest heart rate',xlabel='age',ylabel='highest heart rate')

ax0.legend(*scatter0.legend_elements(),title='Legend')
ax1.legend(*scatter1.legend_elements(),title='Legend')

ax0.axhline(over50['chol'].mean(),linestyle='--')
ax1.axhline(over50['thalach'].mean(),linestyle='--')

print(f"Mean of cholestrol is {over50['chol'].mean()}")
print(f"Mean of highest heart rate is {over50['thalach'].mean()}")
```

```python
fig,(ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10))

scatter0 = ax0.scatter(x=over50['age'],
           y=over50['chol'],
           c=over50['target'],
           cmap='summer')

scatter1 = ax1.scatter(y= over50['thalach'],
            x= over50['age'],
           c= over50['target'],
           cmap='summer')

ax0.set(title='Relation between age and cholestrol levels',
       xlabel='age',
       ylabel='cholestrol')

ax1.set(title='Relation between age and max heart rate',
       xlabel='age',
       ylabel='max heart rate')

ax0.axhline(over50['chol'].mean(),linestyle='-.')
ax1.axhline(over50['thalach'].mean(),linestyle='-.')

ax0.legend(*scatter0.legend_elements(),title='Legend')
ax1.legend(*scatter1.legend_elements(),title='Legend')

plt.suptitle("Heart disease analysis",fontsize=18,y=0.95);
```
